```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# 1. Configuración de parámetros
num_registros = 1000
ahora = datetime.now()
fecha_inicio = ahora - timedelta(days=395)  # Aprox 13 meses atrás

# 2. Listas de datos base
ciudades = ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Bucaramanga']
canales = ['Web', 'App', 'Redes']
productos = {
    'Electrónica': [('Laptop Pro', 1200, 800), ('Monitor 4K', 350, 200), ('Teclado Mec', 80, 45)],
    'Accesorios': [('Mouse Optico', 25, 10), ('Headset', 60, 30), ('Cable HDMI', 15, 5)],
    'Muebles': [('Silla Gamer', 250, 150), ('Escritorio L', 180, 100)]
}

data = []

for i in range(1, num_registros + 1):
    dias_aleatorios = random.randint(0, 395)
    fecha_obj = fecha_inicio + timedelta(days=dias_aleatorios)

    # ERROR 1: Formatos de fecha inconsistentes
    if i % 10 == 0:
        fecha = fecha_obj.strftime('%m/%d/%Y')
    else:
        fecha = fecha_obj.strftime('%Y-%m-%d')

    cat = random.choice(list(productos.keys()))

    # ERROR 2: Inconsistencia en nombres de categorías
    if i % 15 == 0:
        cat = "Elec"

    prod_info = random.choice(productos[list(productos.keys())[0] if i % 15 == 0 else cat])

    # ERROR 3: Outliers y Nulos
    cantidad = random.randint(1, 10)
    if i % 20 == 0:
        cantidad = -5

    precio = prod_info[1]
    if i % 25 == 0:
        precio = None

    costo = prod_info[2]

    # ERROR 4: Tipo incorrecto
    if i % 50 == 0:
        costo = "ERROR_SYS"

    data.append([
        i, fecha, random.choice(ciudades), random.choice(canales),
        prod_info[0], cat, cantidad, precio, costo, random.randint(1, 5)
    ])

# 3. Crear DataFrame
columnas = ['OrderID', 'Fecha', 'Ciudad', 'Canal', 'Producto', 'Categoria', 'Cantidad', 'PrecioUnitario', 'CostoUnitario', 'CalificacionCliente']
df_ventas = pd.DataFrame(data, columns=columnas)

# 4. Mostrar todos los datos en consola
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print(df_ventas.to_string(index=False))

# 5. Guardar en Excel
df_ventas.to_excel("ventas_ecommerce.xlsx", index=False)

print("\nArchivo guardado como ventas_ecommerce.xlsx")
```
